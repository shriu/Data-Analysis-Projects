```python

df_tracks['release_date'].unique()
```

```python
#converting release_date to dateand time format
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'])
df_tracks.info()
```

```python
#df_tracks.set_index('release_date' , inplace=True)
```

```python
#converting duration_ms from ms to mins
df_tracks['duration_ms'] = df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.head(2)
```

```python
#Renaming duration_ms column to duration :
df_tracks.rename(columns = {'duration_ms' : 'duration'} , inplace=True)
df_tracks.head(2)
```

```python
#corr_df = df_tracks.corr(method= 'pearson')
plt.figure(figsize=(15,10))
sns.heatmap(df_tracks.corr(), annot = True, cmap = "RdYlGn")
```

## Observations:
Danceability and valence, spichiness and liveliness,loudness and popularity,valence and energy, energy and loudness all have a favourable association.

```python
sample_df = df_tracks.sample(int(0.005*len(df_tracks)))
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data=sample_df ,x ='valence' , y='danceability', scatter_kws={"color": "blue"}, line_kws={"color": "red"})
sns.set_style(style = "darkgrid")
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Valence',size = 15 )
plt.ylabel('Danceability', fontsize = 15)
```

### Observations :

Valence and danceability have a good correlation with each other.

```python
#showing the correlation between loudness and opularity using regression plot using only a few samples

plt.figure(figsize=(12,6))
sns.regplot(data=sample_df ,x ='popularity' , y='loudness', scatter_kws={"color": "blue"}, line_kws={"color": "red"})
sns.set_style(style = "darkgrid")
plt.xticks(size = 15)
plt.yticks(size=15)
plt.ylabel('Loudness',size = 15 )
plt.xlabel('Popularity', fontsize = 15)
```

### Observations :

The average popularity ranged from 0 to 60% for the loudness between -20 and 0.

```python
#plotting a regression for danceability to popularity
plt.figure(figsize=(12,6))
sns.regplot(data = sample_df , x='popularity' , y='danceability' , scatter_kws={'color' : 'green'} , line_kws={'color' :'red'})
sns.set_style(style = "darkgrid")
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Popularity',size = 15 )
plt.ylabel('Danceability', fontsize = 15)
```

### Observations:
Popularity and danceability have a positive link, but it is not very strong. The transition is not that sudden.

It's as though people dance to tunes regardless of how popular they are.

Even unpopular songs that are not popular are danceable.


```python
#regplot for loudness vs energy
plt.figure(figsize=(12,6))
sns.regplot(data = sample_df , x='energy' , y='loudness' , scatter_kws={'color' : 'blue'} , line_kws={'color' : 'red'})
sns.set_style(style = "darkgrid")
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Energy',size = 15 )
plt.ylabel('Loudness', fontsize = 15)
```

### Observations:
The energy rises in proportion to the loudness.

```python
#regplot for accoustic vs popularity
plt.figure(figsize=(12,6))
sns.regplot(data = sample_df , x='acousticness' , y='popularity' , scatter_kws={'color' : 'blue'} , line_kws={'color' : 'red'})
sns.set_style(style = "darkgrid")
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Energy',size = 15 )
plt.ylabel('Loudness', fontsize = 15)
```

## Observations:
As the acousticness increases, the level of popularity decreases.

```python
#songs with more than 95 popularity
most_popular_songs = df_tracks[df_tracks['popularity']>92].sort_values('popularity', ascending= False)
most_popular_songs.info()
```

```python
most_popular_songs['year'].value_counts()
```

```python
#most popular songs :
plt.figure(figsize=(12,6))
sns.barplot(x='popularity', y = 'name', data = most_popular_songs, palette = ('hls'))
plt.title('Most popular songs', fontweight = 'bold', size = 20)
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Popularity',size = 15 )
plt.ylabel('Songs', fontsize = 15)

plt.show()
```

## Observations:
The songs from the years 2020 and 2021 were the most well-liked.
Of all the most popular songs, Justin Bieber's Peaches was a huge hit.

```python
plt.figure(figsize=(20,10))
sns.barplot(x = 'year', y = 'duration', data = df_tracks )
plt.title("Year vs Duration", fontweight='bold', size=20)
plt.xticks(rotation = "90", size = 13)
plt.yticks(size=15)
plt.xlabel('Year',size = 15 )
plt.ylabel('Duration of Songs', fontsize = 15)

plt.show()
```

## Observations:
After analyzing in depth , the maximum duration of songs were in the 1935s.
It saw a continuous growth from the 1980s where duration peaked  1988 and in 2010.
It similarly abruptly peaked in 1969. 
However the duration decreased from 2013

```python
plt.figure(figsize=(20,10))
sns.lineplot(x = 'year', y = 'duration', data = df_tracks)
plt.title("Year vs Duration", size = 20, fontweight =  'bold')
plt.xticks(size = 15)
plt.yticks(size=15)
plt.xlabel('Year',size = 15 )
plt.ylabel('Duration of Songs', size = 15)

sns.set_style(style="whitegrid")
plt.show()
```

## Observations:
Around 1920s, there was a sharp decline in music duration; however, it steadily rose and peaked between the 1930s and 1940s.
